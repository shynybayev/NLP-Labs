Text cleaning

```
import re
result = []
def clean_text(txt):
  txt = txt.strip()
  txt = txt.replace('\n', ' ')
  txt = txt.replace(' .', '.')
  txt = txt.replace(' ,', ',')
  txt = txt.replace(' ;', ';')
  txt = txt.replace(' :', ':')
  txt = txt.replace(' ?', '?')
  txt = txt.replace(' !', '!')
  txt = txt.replace(' )', ')')
  txt = txt.replace('( ', '(')
  txt = txt.replace('« ', '«')
  txt = txt.replace(' »', '»')
  txt = txt.replace('&quot;', '"')
  return txt
```

Making frequency table and create a dictionary for the word frequency table


```
import nltk
nltk.download('punkt')

words = word_tokenize(text)
```

```
freqTable = dict()
for word in words:
    word = word.lower()
    if word in stopWords:
        continue
    if word in freqTable:
        freqTable[word] += 1
    else:
        freqTable[word] = 1
```

Assigning a score to every sentence


```
sentences = sent_tokenize(text)
sentenceValue = dict()
```

```
for sentence in sentences:
    for wordValue in freqTable:
        if wordValue[0] in sentence.lower():
            if sentence in words:
                sentenceValue[sentence] += freqTable[wordValue]
            else:
                sentenceValue[sentence] = freqTable[wordValue]
```

Finding the average score of a sentence

```
sumValues = 0
for sentence in sentenceValue:
    sumValues += len(sentenceValue[sentence])

# Average value of a sentence from original text
average = int(sumValues/ len(sentenceValue))
```

To apply our threshold and store our sentences in order into our summary.



```

summary = ''
for sentence in sentences:
     if (sentence in sentenceValue) and (sentenceValue[sentence] > (1.5 * average)):
          summary += " " + sentence
          break;
          
result = [summary]
```
